```python
import numpy as np
import matplotlib.pyplot as plt

#w= 10**(-3)
w= 10**(3)
c= 10**(-6)

def phi_t(t):
    return w*t

def r_t(t):

    return(c/2)*(np.exp(w*t)+np.exp(-w*t))



def r_phi(phi):
    return(c/2)*(np.exp(phi)+np.exp(-phi))

tiempos=np.linspace(0,5,100000)
```

```python

plt.plot(tiempos,phi_t(tiempos))
plt.xlabel("Tiempo (s)")
plt.ylabel("Phi (rad)")
plt.title("Phi vs Tiempo")
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
phis= phi_t(tiempos)

ax.plot(phis, r_phi(phis), linewidth=2)

# Customize the plot
ax.set_rlabel_position(135)
ax.set_title("r vs phi", pad=10)
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
phis= np.linspace(0, 6 * np.pi, 1000)

ax.plot(phis, r_phi(phis), label='Spiral', linewidth=2)

# Customize the plot
ax.set_rlabel_position(135)
ax.set_title("Spiral Plot")
ax.legend()
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

ax.plot(tiempos, r_t(tiempos), linewidth=2)

# Customize the plot
ax.set_rlabel_position(135)
ax.set_title("r vs t", pad=10)
```
